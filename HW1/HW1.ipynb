```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from scipy.sparse import csr_matrix
import joblib
```

```python
# Reading Parquet file with pandas and pyarrow
df1 = pd.read_parquet('yellow_tripdata_2023-01.parquet', engine='pyarrow')
df1
```

```python
# Reading Parquet file with pandas and pyarrow
df2 = pd.read_parquet('yellow_tripdata_2023-02.parquet', engine='pyarrow')
df2
```

```python
df = df1.copy()
df['duration'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60
df['duration'].describe()
```

# Q4

```python
df1['PULocationID'].isna().sum()
```

```python
df1['DOLocationID'].isna().sum()
```

```python
data_dicts = df[['PULocationID', 'DOLocationID']].astype(str).to_dict(orient='records')
data_dicts[:5]
```

```python
# Fit a dictionary vectorizer
vec = DictVectorizer(sparse=False)
feature_matrix = vec.fit_transform(data_dicts)
matrix = feature_matrix.astype(np.float32)
del feature_matrix
```

```python
print("Feature Matrix:")
print(matrix.shape)
```

```python
# Because matrix contains many zeros, using a sparse matrix format can save a lot of memory.
X_sparse = csr_matrix(matrix)
X_sparse
```

# Q5

```python
# Target variable
y = df['duration']

# Train a plain linear regression model with default parameters
model = LinearRegression()
model.fit(X_sparse, y)
```

```python
joblib.dump(model, 'linear_regression_model.joblib')
```

```python
# Make predictions on the training data
y_pred = model.predict(X_sparse)

# Calculate the RMSE
rmse = np.sqrt(mean_squared_error(y, y_pred))

print("Root Mean Squared Error (RMSE):", rmse)
```
